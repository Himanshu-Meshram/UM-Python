# **Univariate Plots**

The term univariate analysis refers to the **analysis of one variable**. You can remember this because the prefix "uni" means "one". The purpose univariate analysis is to understand the distribution of values for a single variable.

In histogram, the x axis represents the bin ranges and y axis represents the information about the frequency of the data.

> # **Pie Chart**

A pie Chart is a circle divided into sectors that each represent a proportion of the whole. It is often used to show percentage, where the sum of the sectors equals 100%. They are often used to represent sample data - with data points belonging to a combination of different categories. Each of these categories is represented as a "slice of the pie". The size of each sliceis directly proportional to the number of data points that belong to a particular category.

> # **2. Scatter plot**
A scatter plot is a type of data visualization that shows the **relationship between different variables**. This data is shown by placing various data points between an X and Y axis. Scatter plot can also be known as scatter diagrams or X-Y graphs, and the point of using one of these is to determine id there are patterns or correlations between teo variables.
 * Both columns should be numeric.

 **When to use :** Use a scatter plot to determine or not two variables have a relationship or correlation.

 **When not to use :** Avoid a scatter plot when your data is not at all related or you have too large dataset. (For data set with hundreds of thousand or millions of records, a scatterplot will be to dense, so we need a different way to visualize the relationship.)

> # **3. Line Plot**

 A line chart, also known as a line graph or curve chart, is a graphical representation used to **display data points connected by straight lines**. This type of chart is particularly useful for **visualizing trends, changes, and relationships in data** over a continuous interval, often time. Each data point on the chart represents a value associated with a specific category or time point. The connecting lines between these points allow viewers to discern patterns and variations in the data with ease.
