# **TASK 2 -** Visualise the distribution of ratings and compensations.

We will now visualise the dsitributions of employee ratings and compensations.

we will create the following plots:

* A scatter plot of employee age and employee salary, healthy_eating active_lifestyle, healthy_eating and salary.
* A count plot of Groups.
* A histogram of employee salary.


```python
## create a scatter plot of the "Age" and "Employee Salary" column.

plt.figure(figsize=(10,5))

sns.scatterplot(data=data, x='age',y='salary',color= 'blue',
                edgecolor='black', alpha =0.3)
plt.title('Age vs Salary')                                                              ## the dark dots are the points overlapped on other.
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

**Observations:**

We observe that in general, there is a relationship between age and salary of the employee.

As age increases, the Salary also increases, though there are some outliers present as well

```python
## Create a scatter plot of the "healthy_eating" and "active_lifestyle" column

plt.figure(figsize=(10,5))

sns.scatterplot(data=data, x='healthy_eating',y='active_lifestyle', color='red',
                edgecolor='orange', alpha =0.5)
plt.title('Healthy Eating vs Active Lifestyle')
plt.xlabel('Healthy Eating')
plt.ylabel('Active Lifestyle')
plt.show()
```

**Observations:**

We observe that in general, there is a relationship between healthy_eating and Active Lifestyle of employees.

As healthy_eating increases, the active_lifestyle also increases.


```python
## Create a scatter plot of the "healthy_eating" and "salary" columns.

plt.figure(figsize=(10,5))

sns.scatterplot(data=data, x='healthy_eating',y='salary', color='green',
                edgecolor='white', alpha = 0.5)
plt.title('Healthy Eating vs Salary')
plt.xlabel('Healthy Eating')
plt.ylabel('Salary')
plt.show()



```

**Observations:**

We observe that in general, there is a relationship between healthy_eating and salary of employee.

As healthy_eating increase, the salary also increases.

```python
## Create a Count Plot of the "Groups" Column

plt.figure(figsize=(10,5))

sns.countplot(data=data, x='groups',
              color = 'blue', edgecolor='red', alpha=0.5)
plt.title('Count Plot of Groups')
plt.xlabel('Groups')
plt.ylabel('Count')
plt.show()
```

**Observations:**

We see that most of the employees either belong to Group A or Group B, with Group A having maximum frequency

```python
## Create a Histogram of the "salary" Column

plt.figure(figsize=(10,5))

sns.histplot(data=data, x='salary', color='pink',
             edgecolor='red', alpha=0.8,bins =10)
plt.title('Histogram of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

**Observations:**

We see that the salaries are not uniform, which is not necessarily a discrepancy.

But we do observe that there are employees with salaries on either extremities of the histogram.

Based on the histogram, we can say that the count of highest earning employee is very less.

# **TASK 3 -** Subset the data based on the thresholds

We will now subset the original data frame based on the following conditions:
* Employees with healthy_eating greater than 8.
* Employees with salary less than 1000.
* Employees with healthy_eating greater than 8 and with salary less than 1000.

```python
## Employee healthy_eating is greater than 8

sub1 = data[data['healthy_eating']>8]                                           ## data for the csv file , data for the columns values.
sub1
```

```python
## salary less than 1000

sub2 = data[data['salary']<1000]
sub2
# sub2.count()
```

```python
## COMPARE employee with healthy_eating greater than 8 and with salary less than 1000

sub3 = data[(data['healthy_eating']>8) & (data['salary']<1000)]
sub3
```

**Observations:**

1. The only employee seemingly facing a discerpancy in salary as compared to healthy_eating is employee who is having id = 26.
2. The employee with id having 26 has a salary of 700.
