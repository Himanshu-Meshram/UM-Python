# **STATISTICS**

```python
import math
import numpy as np
```

**POPULATION VARIANCE**
16std_D3.png

```python
##-----##----- To find POPULATION VARIANCE

popu= np.arange(0,50,5)                                                         ## Creating Array from 0-50 with 5 stepsize
# popu                                                                            ## result is the >> (Xi)
                                                                                  ## number of elements >> N
                                                                                    ## left with u


## -----##----- For population MEAN "u"          {center value}


pop_mean = sum(popu)/len(popu)                                                  ## mean >> sum of nums / length of num

# pop_mean                                                                          ## syntax {mean = sum(-variable-)/len(-variable-)}




##------##------ Calculating POP_DEVIATION          {(Xi-u)}
                                                                                    ## calculating the distance between the mean and the desired value.

sq_pop_devi = [(i-pop_mean)**2 for i in popu]                                  ### i >> each element

# sq_pop_deviation



##-----##------ Calculating POPULATION VARIATION

pop_var = sum(sq_pop_devi)/len(popu)                                       ## numerator / denominator

# pop_var

# print('Population Variance : {}'.format(pop_var))
# print('Population_Variance :{}'.format(round(pop_var)))


```

**SAMPLE VARIANCE**

```python
sample = np.arange(0,25,5)                                                        ## Creating Array from 0-50 with 5 stepsize
sample                                                                            ## result is the >> (Xi)
                                                                                    ## number of elements >> n



## -----##----- For population MEAN "u"          {center value}

s_mean = sum(sample)/len(sample)
s_mean



##------##------ Calculating POP_DEVIATION          {(Xi-x_)}

sam_devi = [(i-s_mean)**2 for i in sample]
sam_devi




##-----##------ Calculating SAMPLE VARIATION

sam_var = sum(sam_devi)/(len(sample)-1)

# # print('sam_var : {}'.format(sam_var))
# print('sam_devi : {}'.format(round(sam_var)))                                   ## syntax {print('-variable-: {}'.format(round(std,2))}


```
