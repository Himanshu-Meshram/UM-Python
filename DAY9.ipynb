# **Top Commands**



*   Creating New, Renaming & Droping Columns.
*   Apply Method, Sort Index, Sort Values.



```python
import pandas as pd
data = pd.DataFrame({'Name' : ['Ron','Willy','Rega','Mamba','Zeri','Mortal','Viper'],
              'Age' : [22,28,33,26,34,31,30],
              'Uni' : ['BHU','MSU','JNU','DU','SPPU','DU','JNU']})
# data


                                                                                ## Creating New Columns.

                                                                                ##      Syntax  data['-Column-head-']=['values ....']

data['SUBJECT']=['WMCom','WMCom','WMCom','WMCom','WMCom','WMCom','WMCom']
data['City'] = ['Mumbai','Delhi','Pune','Assam','Amritsar','Navi Mumbai','Guwahati']
data



######--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------


                                                                                ##  Renaming  Columns

data.columns = ['NAME','AGE','UNIVERSITY','SUBJECT','CITY']                                   #       syntax    {-variable_name-.columns = ['values...']}
# data.columns = ['NAME','AGE(Yrs)','UNIVERSITY','SUBJECT','CITY']
# data



######--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

                                                                                ## Droping Columns


# data.drop(['AGE'],axis=1)                                                     ## Temporary Dropping columns
# data.drop(['UNIVERSITY'],axis=1)                                                                 ##              syntax   {-variable_name-.drop(['-column1-'],axis=1)}

# data

                                                                                ## Permanently Droping Columns


# data.drop(['AGE'],axis=1,inplace=True)                                                           ##              syntax   {-variable_name-.drop(['-column1-'],axis=1,inplace=True)}
# data



######--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------



data['MARKS']=[26,24,19,22,20,23,27]                                                                ##    syntax    {-variable_name-['-column1-']=['-values.....-']}
# data


                                                                                ## Apply Method


#                        ##         syntax     {-variable_name-['-define_column-']=data['accessing_column'].apply(lambda -arithematical_calc.-)}



data['PERCENTAGE']=data['MARKS'].apply(lambda x:(x/30)*100)
data['RETIREMENT']=data['AGE'].apply(lambda x:65-x)
data



                                                                                ## To show selected columns
# selected_columns = ['NAME','AGE']                                                     ##              syntax {selected_columns = ['columnss...']}
# data[selected_columns]





 ######--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------



data.index = [22,3,44,12,62,21,16]                                                               ## assigning random indexing
data



                                                                                          ##  Ascending Order
# data.sort_index()                                                             ##          syntax    {-variable_name-.sortindex() }

                                                                                          ## Descending Order
# data.sort_index(ascending=False)                                                ##            syntax    {-variable_name-.sortindex(ascending=False) }



######--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

data


data.sort_values('MARKS')
data.sort_values('PERCENTAGE')
### data.sort_values('MARKS','PERCENTAGE')

# data.sort_values('MARKS',ascending=False)
# data.sort_values('RETIREMENT',ascending=False)





```
