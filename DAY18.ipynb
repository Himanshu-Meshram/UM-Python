# **MULTIVARIATE PLOTS**

Multivariate analysis (MVA) is a Statistical procedure for analysis of dat involving **more than one type of (variable/column)** measurement or observation. It may also solving problems where more than one dependent variable is analyzed simulataneously with other variables.



> # **1. Stacked Bar Plot**

A stacked chart is a form of bar chart that **shows the composition and comparison of a few variables**, either relative or absolute over time. Also called as stacked bar or column chart, they look like a series of columns or bars that are **stacked on top of each other.**

> # **2. Multiple Scatterplot**

A 3D scatter plot allows the visualization of multivariate data. This scatter plot **takes multiple scalar variables** and uses them for different axes in phase space. The different variables are combined to form coordinated in the phase space and they are displayed using glyphs and **colored** using another scalar variable.

> # **Donut Chart**

```python
!pip install palettable                                                             ## to use Pastel1_7 lib, we need to install palettable.

import matplotlib.pyplot as plt                                                     ## importing matplotlib
from palettable.colorbrewer.qualitative import Pastel1_7                            ## to import Pastel1_7 Library.

                                                                                    ## it is used to provide colors for the charts and sections in it

##-----##------ Create data

names = ['Sec-A','Sec-B','Sec-C','Sec-D','Sec-E']
size = [12,45,32,55,10]



##-----##------ Create a circle at the center of the plot

in_circle = plt.Circle((0,0),0.7, color='white')



##-----##----- Main pie chart

from palettable.colorbrewer.qualitative import Pastel1_7
plt.pie(size, labels = names, colors = Pastel1_7.hex_colors)                        ## ploting piechart



##-----##----- adding center circle in the piechart
p = plt.gcf()                                                                   ## gcf >> get current figure
p.gca().add_artist(in_circle)                                                   ## gca >> get current artist

plt.show()


```

> # **Tree Map**

* used to show the distribution of the values. in terms of square and rect.

```python
!pip install squarify
import matplotlib.pyplot as plt
import squarify                                                                   ## squarify is the algorithm used in tree map
import pandas as pd


## Create a DataFrame with sample data

data = pd.DataFrame({'People': [3,12,9,10],                                       ## Creating data
                     'Group':['GRP-A','GRP-B','GRP-C','Grp-D']})


##-----##----- Plotting

squarify.plot(sizes=data['People'],label=data['Group'],alpha=.8)                     ## syntax >> squarify.plot(sizes=-data_variable-['-vaiable-'],label=-data_variable-['-variable-'],alpha=.8)
plt.axis('off')                                                                     ## axis >> off

plt.show()


#################______PRACTICE______________


### data = pd.DataFrame({'People': [8,3,4,2],                                       ## Creating data
###                      'Group':['GRP-A','GRP-B','GRP-C','Grp-D']})


# ##-----##----- Plotting

### squarify.plot(sizes=data['People'],label=data['Group'],alpha=.8)                     ## syntax >> squarify.plot(sizes=-data_variable-['-vaiable-'],label=-data_variable-['-variable-'],alpha=.8)
### plt.axis('off')                                                                     ## axis >> off

### plt.show()
```

> # **Background Map**

```python
import pandas as pd

###-----Creating a Dataframe with dots to show on the map.


data = pd.DataFrame({'lon' :[-58,2,145,30.32,-4.03,-73.57,-36.82,-38.5],
                     'lat' :[-34,49,-38,59.93,5.33,45.35,-1.29,-12.97],
                     'name': ['Agra','Delhi','Goa','MP','Nepal','Nagaland','Sikkim','HP'],
                     'value': [10,12,40,70,23,43,100,43]},
                    dtype=str)

data



!pip install folium                                                             ##--------##---------- FOLIUM IS USED TO DOWNLOAD THE WORLD MAP.............
import folium



## make an empty map

m = folium.Map(location=[20,0], tiles='OpenStreetMap',zoom_start=2)                ### syntax >> -variable- = folium.Map(location=[x,y],tiles="OpenStreetMap",zoom_start=2)

##----Show
m


##---- to add the data into the map

for i in range(0,len(data)):
  folium.Marker(                                                                ## adding markers at the locations
      location=[data.iloc[i]['lat'],data.iloc[i]['lon']],                       ## giving location from the data frame
      popup=data.iloc[i]['name'],
  ).add_to(m)                                                                   ## adding to the empty map created


m


```
